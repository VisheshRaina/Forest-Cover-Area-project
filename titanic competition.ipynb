```python
import pandas as pd
#import quandl, 
import math, datetime
import numpy as np
from sklearn import preprocessing,  svm
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from matplotlib import style
```

```python
data=pd.read_csv('train.csv')
```

```python
data
```

```python
data['Sex'].replace('female',0,inplace=True)
data['Sex'].replace('male',1,inplace=True)
data['Embarked'].replace('S',0,inplace=True)
data['Embarked'].replace('C',1,inplace=True)
data['Embarked'].replace('Q',2,inplace=True)
##### Male is 1 & female is 0
## S is 0, C is 1, Q is 2
```

```python
data.head(7)
```

```python
new_data=data.interpolate(method='quadratic')
new_data.head(19)
```

```python
new_data.drop(['Name','Cabin','Survived','Ticket'],axis=1,inplace = True)
```

```python
X=new_data
X.head()
```

```python
labels_t=data.T[1:2]
y=labels_t.T
y=y.astype('int')
y.head()
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.3)
```

```python
#y_train.head()
X_train.head()
```

```python
nx=X.shape[1]
print (nx)             
m=X.shape[0]
print(m)              
```

```python
classifier=LogisticRegression(fit_intercept=True, max_iter=2000).fit(X_train,y_train)
```

```python
classifier.score(X_val,y_val)             
    
```
