# choosing a K-value


```python
error_list=[]
for i in range(1,8):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    predi=knn.predict(X_val)
    pred_i=predi.reshape(5292,1)
    error_list.append(np.mean(pred_i!=y_val))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,8),error_list,linestyle='dashed',marker='o',markerfacecolor='blue',markersize=12)
plt.xlabel('K')
plt.ylabel('Error rate')
```
